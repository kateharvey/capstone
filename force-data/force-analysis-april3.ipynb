```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

### Import data

```python
force_old = pd.read_csv('./apr3/apr3-force-old.csv')
pressure_old = pd.read_csv('./apr3/apr3-pressure-old.csv')
force_new = pd.read_csv('./apr3/apr3-force-new.csv')
pressure_new = pd.read_csv('./apr3/apr3-pressure-new.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(pressure_old['Air Pressure'], label='Pressure')
ax1.plot(pressure_old['PWM'], label='PWM')
ax1.set(xlim=(0, 1600))
ax1.set(title='Old Pumps', xlabel='Time', ylabel='Air Pressure (mbar)')
ax1.legend()

ax2.plot(pressure_new['Air Pressure'], label='Pressure')
ax2.plot(pressure_new['PWM'], label='PWM')
ax2.set(title='New Pumps', xlabel='Time', ylabel='Air Pressure (mbar)')
ax2.legend()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(force_old['Force (N)'], label='Force')
# ax1.plot(pressure_old['PWM'], label='PWM')
ax1.set(xlim=(0, 1050))
ax1.legend()

ax2.plot(force_new['Force (N)'], label='Force')
# ax2.plot(force_new['PWM'], label='PWM')
ax2.set(xlim=(0, 1500))
ax2.legend()
plt.show()
```
