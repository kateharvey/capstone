```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in the data
syringe_data = pd.read_csv('pasco-data-march1/syringe-inflation.csv')
syringe_data.head()
```

```python
%matplotlib inline

# plot the data
fix, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(syringe_data["Time (s) Run #1"], syringe_data["Force (N) Run #1"])
ax2.plot(syringe_data["Time (s) Run #2"], syringe_data["Force (N) Run #2"])
ax1.set_title('Force vs. Time  Trial 1')
ax2.set_title('Force vs. Time  Trial 2')
ax1.set_xlabel('Time (s)')
ax2.set_xlabel('Time (s)')
ax1.set_ylabel('Force (N)')
ax2.set_ylabel('Force (N)')
plt.show()
```
